# IMDB Sentiment Analysis

## Prepare Data

```python
# See https://github.com/fchollet/keras/blob/master/examples/imdb_lstm.py
from keras.datasets import imdb
from keras.preprocessing import sequence
import numpy as np

max_features = 20000
maxlen = 150  # cut texts after this number of words (among top max_features most common words)

print('Loading data...')
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
print(len(x_train), 'train sequences')
print(len(x_test), 'test sequences')

word_index = imdb.get_word_index()

print('Pad sequences (samples x time)')
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
```

## Simplest LSTM model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

simpleModel = Sequential([
    LSTM(64, input_shape=(150, 1,)),
    Dense(1, activation='sigmoid')
])

simpleModel.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history = simpleModel.fit(np.expand_dims(x_train, axis=2), y_train,
                        validation_data=(np.expand_dims(x_test, axis=2), y_test),
                        epochs=20, batch_size=2500)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plotLoss(hist):
    # summarize history for loss
    plt.plot(hist.history['loss'])
    plt.plot(hist.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    
plotLoss(history)
```

## More Layers + Batch Normalization

```python
from keras.layers.normalization import BatchNormalization

moreLayers = Sequential([
    LSTM(32, input_shape=(150, 1,), return_sequences=True),
    BatchNormalization(),
    LSTM(32),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])

moreLayers.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
model2Hist = moreLayers.fit(np.expand_dims(x_train, axis=2), y_train,
                        validation_data=(np.expand_dims(x_test, axis=2), y_test),
                        epochs=20, batch_size=2000)
plotLoss(model2Hist)
```

## Add embedding layer

```python
import os

embeddings_index = {}
f = open(os.path.join('../datasets', 'glove.42B.300d.txt'))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
EMBEDDING_DIM = 300

embedding_matrix = np.zeros((len(word_index) + 1, EMBEDDING_DIM))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector
```

```python
from keras.layers.normalization import BatchNormalization
from keras.layers import Embedding

withEmbedding = Sequential([
    Embedding(len(word_index) + 1,
            EMBEDDING_DIM,
            weights=[embedding_matrix],
            input_length=maxlen,
            trainable=False),
    LSTM(64, dropout=0.5, return_sequences=True),
    BatchNormalization(),
    LSTM(16, dropout=0.5),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])

withEmbedding .compile(loss='binary_crossentropy', optimizer='adam')
model3Hist = withEmbedding.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=20, batch_size=2000)
plotLoss(model3Hist)
```

## Tuning

```python
from keras.layers.normalization import BatchNormalization
from keras.layers import Dense
from keras.layers import Embedding

tuned = Sequential([
    Embedding(len(word_index) + 1,
            EMBEDDING_DIM,
            weights=[embedding_matrix],
            input_length=maxlen,
            trainable=False),
    LSTM(256, dropout=0.5, return_sequences=True),
    BatchNormalization(),
    LSTM(128, dropout=0.5),
    BatchNormalization(),
    Dense(16, activation='relu'),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])

tuned.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
model4Hist = tuned.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=20, batch_size=1500)
plotLoss(model4Hist)
```

```python
tuned.layers[0].trainable = True
tuned.optimizer.lr = 1e-5
model4ExtraTrainingHist = tuned.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=10, batch_size=1500)
plotLoss(model4ExtraTrainingHist)
```

## How much better can we do?

```python
tuned.optimizer.lr = 1e-6
model4EvenMoreTraining = tuned.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=20, batch_size=1500)
plotLoss(model4EvenMoreTraining )

# I've tried training beyond this, but it doesn't go beyond about 85% accuracy
```

## Simplification, and some more tuning

```python
from keras.optimizers import Adam

simplified = Sequential([
    Embedding(len(word_index) + 1,
            EMBEDDING_DIM,
            weights=[embedding_matrix],
            input_length=maxlen,
            trainable=False),
    LSTM(80, dropout=0.5, return_sequences=True),
    BatchNormalization(),
    LSTM(80, dropout=0.5),
    BatchNormalization(),
    Dense(16, activation='relu'),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])

simplified.compile(loss='binary_crossentropy', optimizer=Adam(lr=1e-3), metrics=['binary_accuracy'])
model5Hist = simplified.fit(x_train, y_train,
                            validation_data=(x_test, y_test),
                            epochs=20, batch_size=2500)
plotLoss(model5Hist)
```

```python
simplified.optimizer.lr = 1e-4
plotLoss(simplified.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=20, batch_size=2500)
        )

simplified.optimizer.lr = 1e-5
plotLoss(simplified.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=20, batch_size=2500)
        )
```

```python
simplified.layers[0].trainable = True
plotLoss(simplified.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=10, batch_size=2500)
        )
```

```python
simplified.optimizer.lr = 1e-6
plotLoss(simplified.fit(x_train, y_train,
                        validation_data=(x_test, y_test),
                        epochs=25, batch_size=2500)
        )
```

## Add a Convolutional Layer

```python
embedding_matrix[0].shape
```

```python
from keras.layers.convolutional import Conv1D
from keras.layers.pooling import MaxPooling1D

n_gram = 3

with_conv = Sequential([
    Embedding(len(word_index) + 1,
            EMBEDDING_DIM,
            weights=[embedding_matrix],
            input_length=maxlen,
            trainable=False),
    Conv2D(20, kernel_size=(1, EMBEDDING_DIM), strides=n_gram, activation='relu'),
    MaxPooling2D(pool_size=(np.floor(maxlen / n_gram), EMBEDDING_DIM)),
    BatchNormalization(),
    LSTM(80, dropout=0.5),
    BatchNormalization(),
#     Dense(16, activation='relu'),
#     BatchNormalization(),
    Dense(1, activation='sigmoid')
])

with_conv.compile(loss='binary_crossentropy', optimizer=Adam(lr=1e-3), metrics=['binary_accuracy'])
model6Hist = with_conv.fit(x_train, y_train,
                            validation_data=(x_test, y_test),
                            epochs=20, batch_size=2500)
plotLoss(model6Hist)
```
